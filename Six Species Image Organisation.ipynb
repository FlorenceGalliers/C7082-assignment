```
import os, shutil
```

```
# The path to the directory where the original dataset was uncompressed
original_dataset_dir = '/content/drive/MyDrive/Harper/3-C7082/Assignment/C7082-assignment/all-data'

# The directory where we will store our smaller dataset
base_dir = '/content/drive/MyDrive/Harper/3-C7082/Assignment/C7082-assignment/6-data'
#os.mkdir(base_dir)
```

Black Grass Directories and Images

```
# Directory with our training bgs pictures
train_bgs_dir = os.path.join(train_dir, 'bgs')
os.mkdir(train_bgs_dir)
# Directory with our validation bgs pictures
validation_bgs_dir = os.path.join(validation_dir, 'bgs')
os.mkdir(validation_bgs_dir)
# Directory with our test bgs pictures
test_bgs_dir = os.path.join(test_dir, 'bgs')
os.mkdir(test_bgs_dir)

#original bgs data
original_bgs_dir = '/content/drive/MyDrive/Harper/3-C7082/Assignment/C7082-assignment/all-data/black-grass'

# Copy first 200 bgs images to train_bgs_dir
fnames = ['bgs_{}.png'.format(i) for i in range(200)]
for fname in fnames:
    src = os.path.join(original_bgs_dir, fname)
    dst = os.path.join(train_bgs_dir, fname)
    shutil.copyfile(src, dst)

# Copy next 25 bgs images to validation_bgs_dir
fnames = ['bgs_{}.png'.format(i) for i in range(200, 225)]
for fname in fnames:
    src = os.path.join(original_bgs_dir, fname)
    dst = os.path.join(validation_bgs_dir, fname)
    shutil.copyfile(src, dst)
    
# Copy next 25 bgs images to test_bgs_dir
fnames = ['bgs_{}.png'.format(i) for i in range(225, 250)]
for fname in fnames:
    src = os.path.join(original_bgs_dir, fname)
    dst = os.path.join(test_bgs_dir, fname)
    shutil.copyfile(src, dst)
```
